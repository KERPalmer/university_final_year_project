```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample gradient data (e.g., representing fluctuations in stock price gradients)
np.random.seed(0)
gradients = np.random.normal(0, 1, 100)  # 100 random gradient values as example
```

```python
# Parameters for gradient smoothing
alpha = 0.2  # Smoothing factor (higher alpha means more smoothing)

# Function to smooth gradients using Exponential Moving Average (EMA)
def smooth_gradients(gradients, alpha=0.1):
    smoothed_gradients = []
    ema = 0  # Start with an initial EMA value of 0
    for gradient in gradients:
        ema = alpha * gradient + (1 - alpha) * ema  # EMA formula
        smoothed_gradients.append(ema)
    return smoothed_gradients

# Smooth the gradients
smoothed_gradients = smooth_gradients(gradients, alpha)

# Plotting to visualize the effect of gradient smoothing
plt.figure(figsize=(12, 6))
plt.plot(gradients, label="Original Gradients", color="blue", linestyle="--")
plt.plot(smoothed_gradients, label="Smoothed Gradients (EMA)", color="red")
plt.xlabel("Time Steps")
plt.ylabel("Gradient Value")
plt.legend()
plt.title("Gradient Smoothing with Exponential Moving Average (EMA)")
plt.show()
```
